### AUC

```python
recal = [(1+cut_off)/2*(1-cut_off)*2 for cut_off in np.linspace(0,1,101)]
print(f"Test recall: {recal[50]}")

fpr = [(1-cut_off)/2*(1-cut_off)*2 for cut_off in np.linspace(0,1,101)]
print(f"Test recall: {fpr[50]}")

plt.plot(fpr, recal)
```

$x = (1-cut\_off)/2*(1-cut\_off)*2 = 1-2cut\_off+cut\_off^2$

$y = (1+cut\_off)/2*(1-cut\_off)*2 = 1-cut\_off^2$



$cut\_off = \sqrt{1-y}$

$x = 1-2\sqrt{1-y}+1-y$


$x = 1-2\sqrt{1-y}+1-y$

$0 = 1 - x -2a+a^2$

$\Delta = 4 - 4(1-x)=4x$

$a_0 = \frac{2\pm \sqrt{4x}}{2}=1\pm \sqrt{x}$, because $a\in(0,1)$, $a_0 =1- \sqrt{x}$

$\sqrt{1-y} = 1- \sqrt{x}$

$1-y = 1- 2\sqrt{x}+x$

$y = 2\sqrt{x}-x$

```python
#Check
x=np.linspace(0,1,101)
y=2*np.sqrt(x)-x
plt.fill_between(x, [0 for _ in y], y, hatch='//', facecolor='blue', alpha=0.5)
plt.plot(x, y)
```

# Beta distribution

```python
import seaborn as sns
a=1/5
x = np.random.beta(a,a,1000)
sns.distplot(x)
plt.show()
counts, bin_edges = np.histogram(x, bins=50)
counts
```

```python
fig, ax = plt.subplots(1,3,figsize=(12,4))
ax[1].plot([.25,.75],[.25,.75], color="blue")
ax[1].plot([.75,.25],[.25,.75], color="blue")
ax[1].plot([0,1],[.5,.5], color="blue")
ax[1].set_xlim(-0.02,1.02)
ax[1].set_ylim(-0.02,1.02)

ax[1].axis('off')
box = ax[1].get_position()
ax[1].set_position([box.x0+box.width*0.4, box.y0, box.width * 0.2, box.height])
box = ax[0].get_position()
ax[0].set_position([box.x0, box.y0, box.width * 1.4, box.height])
box = ax[2].get_position()
ax[2].set_position([box.x0-box.width*0.4, box.y0, box.width * 1.4, box.height])


ax[0].plot([0,1/2,1],[1,1/2,1], color="green")
ax[0].set_xlabel("Observed")
ax[0].set_ylabel("Predicted")
ax[0].fill_between([0,.5,1],[0, 0, 0], [1, .5, 1], hatch='//', facecolor='green', alpha=0.5)
ax[0].set_xlim(-0.02,1.02)

ax[2].bar(np.linspace(0,1,50), counts, color="red", width=1/50)

plt.show()


i = 1
expected = 0
for i in range(len(counts)):
    prob = (2 * i + 1) / 100
    prob = 1 - prob if i<len(counts)/2 else prob
    expected += prob * counts[i]
print(f"Accuracy: {expected/np.sum(counts)}")
```
